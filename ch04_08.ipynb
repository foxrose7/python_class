### ch04-08  원그래프

원 그래프는 전체에 대한 각 용소의 비율을 추출하고, 추출한 비율에 따라 원형을 부체꼴로 분할한 그래프이다. 원 그래프는 각요소의 비율을 비교할 때 이용된다. 

#### 원그래프 그리기

Matplotlib에서는 Axes.pie()에서드를 사용한다. 다음은 코드에서는 pie()메서드의 제1인수에 요소의 값을 부여해서 그리고 있다. 데이터로 리스트형 오브젝트를 이용할 수 있다. 

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')

labels = ['자전거', '버스', '차']
sizes = [25, 40, 35]

fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

#원 그리프 그리기
ax.pie(sizes, labels = labels)

plt.show()
```

Matplotlib의 초기 설정에서는 도수법으로 0도의 위치에서 반시계 방향으로 요소를 그려간다. 중심좌표는 (0,0), 반경은 1이다. 이미지화 하기 쉽게 원 그래프와축의 관계를 명확하게 한 그림을 다음에 나타낸다. 다음 코드에서는 인수 radius에 0.9를, 인수 frame에 True를 설정해서, 축과 함께 그래프를 그리고 있다. 

```python
fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

ax.pie(sizes, labels=labels, radius=0.9, frame=True)
ax.text(-0.3, 0, '(0, 0)',fontsize=9)

plt.show()
```

#### 원그래프 서식 설정

원그래프에서 설정 가능한 서식은 다음과 같다. 

| 항목            |    설명           |
| :------------: | :----------------|
| explode        | 각 요소를 분리하여 표시하는 경우에 설정한다. 리스트형 또는 튜플형을 지정한다. 예를 들어 요소가 4개 있고 3번째 요소를 분리하고 싶은 경우에는 [(0, 0, 0.5, 0)]과 같이 지정한다. |
| labels         | 레이블을 표시한다. 리스트형 또는 튜플형을 지정한다. |
| colors         | 각 요소의 색을 설정한다. 리스트형 또는 튜플형을 지정한다. |
| autopct        | 수치 레이블 서식 설정을 한다. 표시 형식은 문자열로 지정한다.| 
| pctdistance    | 수치 레이블의 위치를 지정한다. 수치열로 지정한다. 수치는 각 요소의 중심부터의 거리가 되고, explode를 설정하고 있는 경우에는 이 거리도 가산된다. |
| shadow         | 배경의 표시, 비표시를 설정한다. 논리값으로 설정한다.|
| labeldistance  | 레이블의 위치를 설정한다. 수치열로 지정한다. 수치는 각 요소의 중심부터의 거리가 되고, explode를 설정하고 있는 경우에는 이 거리도 가산된다. |
| startangle     | 시작 각도를 설정한다. 단위는 도수법으로 수치형으로 지정한다. |
| radius         | 반경을 설정한다. 수치열로 지정한다(초기설정: 1).|
| counterclock   | 표시순서를 설정한다. 논리값으로 설정한다. True: 반시계 방향(초기 설정). False: 시계 방항|
| wedgeprops     | 각 요소의 서식을 설정한다. 서식을 등록한 사전을 지정한다. |
| textprops      | 텍스트의 서식을 설정한다. 서식을 등록한 사전을 지정한다. |
| center         | 원그래프의 중심 좌표를 설정한다.  튜플형을 지정한다. |
| frame          | 축, 테두리선의 유무를 설정한다. 논리값으로 지정한다. |


#### 원그래프의 서식 설정을 하는 경우

- 원그래프 서식 설정

```python
# 원그래프 서식 설정

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# 부채골 서식 설정용 사전
wprops = {
    'edgecolor' : 'black', 
    'linewidth' : 2
}

# 텍스트 서식 설정용 사전
tprops = {
    'fontsize' : 18
}


ax.pie(
    sizes,
    explode=(0.05, 0.0, 0.0),
    labels = labels, 
    autopct = '%1.0f%%',
    pctdistance = 0.5,
    shadow = False,
    labeldistance = 1.35,
    startangle = 90,
    radius = 0.3,
    counterclock = False,
    wedgeprops = wprops,
    textprops = tprops,
    center = (0.5, 0.5),
    frame = True
)

plt.show()
```

#### 원그래프 활용하기

실제의 데이터를 이용해서 그래프를 그린다. 데이터는 anime_genre_top10_pivoted.csv 파일을 이용한다. 


- 데이터 파일 불러오기

```python
import pandas as pd

anime_genre_top10_pivoted_csv = './anime/anime_genre_top10_pivoted.csv'
df = pd.read_csv(anime_genre_top10_pivoted_csv, index_col = 'genre')

df
```

원 그래프는 데이터의 비율을 비교할 때 유용한 그래프이다. 여기서는 Movie와 TV의 총 멤버수 내역을 원그래프로 그려서 장르의 내역 비율을 비교한다. 원그래프는 90도 위치에서 시계 방향으로 내림차순으로 요소를 나열하는 것이 일반적이다. 이 형식을 따라 데이터를 그린다. 

먼저, TV와 Movie의 데이터를 각각 내림차순으로 정렬한 Series를 작성한다. sort_values()의 인수 according에 False를 지정하여 지정한 열이 내림차순으로 소트된다. 

```python
# TV
df_tv = df.sort_values(by='TV', ascending=False)['TV']
# print(df_tv)
#Movie
df_movie = df.sort_values(by='Movie', ascending=False)['Movie']
# print(df_movie
```

- TV 와 Movie 와의 장르 내역 비교를 위한 원 그래프 그리기

```python
fig = plt.figure(figsize = (9, 4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# 컬러 셋트
colors1 =(
    'gold', 'coral', 'plum', 'orchid', 'lightseagreen', 
    'yellowgreen', 'lightskyblue', 'pink', 'cornflowerblue',
    'orangered'
)

colors2 =(
    '#444444', '#555555', '#666666', '#777777', '#888888', 
    '#999999','#AAAAAA', '#BBBBBB', '#CCCCCC', '#DDDDDD'
)

# TV 원그래프
ax1.pie(
    df_tv,
    explode=(0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0.15),       
    labels = df_tv.index,
    autopct = '%1.0f%%',
    colors = colors1, 
    startangle = 90, counterclock=False
)

# Movie 원그래프
ax2.pie(
    df_movie,
    explode=(0, 0, 0, 0, 0, 0, 0.15, 0, 0, 0.15),       
    labels = df_movie.index,
    autopct = '%1.0f%%',
    colors = colors2, 
    startangle = 90, counterclock=False
)


ax1.set_title('TV')
ax2.set_title('Movie')

fig.subplots_adjust(wspace=0.3) # 서브 블록 사이의 공간 조정
plt.show()
```
