### 단축키 모드 표시 하기 : 명령어 모드(Esc)에서 H를 눌러 단축키 목록을 표시할 수 있다. 
### 편리한 단축키
- shift + Enter : 액티브 코드 셀을 실행하고 하나 밑의 셀을 선택한다. 마지막 코드 셀인 경우 실행 후에 새 코드 셀을 도입한다. 이 단축키는 명령 모드에서도 유효하다. 
- ctrl + Enter : 액티브 코드 셀을 실행한다. 이 단축키는 명령어 모드에서도 유효하다(맥에서 ctrl대신 command 키를 사용할 수 없다)
- alt + Enter : 엑티브 코드 셀을 실행하고 하나 밑의 새 코드 셀을 도입한다. 이 단축키는 명령 모드에서도 유효하다
- ctrl + Z : Undo 이다. 직전 편집 내용을 취소한다. 
- tab : 입력중인 코드의 보완 처리를 실행한다. 행의 선두인 경우 인텐드를 도입한다. 
- shift + tab : 툴팁을 표시한다 .
- Esc : 편집 모드에서 명령 모드로 바꾼다. (파란색으로 변경)
- Enter : 명령 모드에서 편집모드로 바꾼다. (녹색으로 변경)

### 명령 모드에서 자주 이용하는 단축키
#### * 특수 문자 입력: command+control+space
- ↑ : 하나 위의 셀을 선택한다. 
- ↓ : 하나 아래의 셀을 선택한다. 
- A : 액티브 코드 셀의 위(Above)에 신규 셀을 추가한다. 
- B : 액티브 코드 셀의 아래(Below)에 신규 셀을 추가한다. 
- Ctrl + S : Notebook 파일을 저장한다. 
- shift : L : 모든 코드 셀의 줄 번호를 보여주거나 숨긴다 (라인넘버(Line number:줄번호) 토글
- d,d : (d를 연속 두 번 입력) 선택 중인 코드 셀을 삭제한다. 
- z : 삭제한 코드 셀을 하나 복원한다. 
- y : 액티브 코드 셀을 Code 타입으로 한다. 코드를 기술하는 타입이다. 
- m : 액티브 코드 셀을 Markdown 타입으로 한다. 
- 0,0 : (0을 두 번 입력) 리스타트 한다. 리스타트 하지 전에 확인 대화상자가 표시된다. 
- p : 명령 팔레트를 연다. 
- h : 단축키 목록을 표시한다. 
- Enter : 선택 중인 코드 셀을 명령 모드에서 편집 모드로 바꾼다.
